```python
class Stack:
    def __init__(self, n):
        self.max_size = n
        self.stack = [None] * self.max_size
        self.top = -1
        
    def push(self, val):
        if self.top == self.max_size - 1:
            print("Stack is full")
        else:
            self.top += 1
            self.stack[self.top] = val
            
    def pop(self):
        if self.top == -1:
            print("Stack is empty")
            return None
        else:
            temp = self.stack[self.top]
            self.stack[self.top] = None
            self.top -= 1
            return temp
            
    def display(self):
        if self.top == -1:
            print("Stack is empty")
            return []
        else:
            return self.stack
            
    def peek(self):
        if self.top == -1:
            print("Stack is empty")
            return None
        else:
            return self.stack[self.top]

ob = Stack(5)
ob.push(10)
ob.push(20)
ob.push(30)
ob.push(40)
print("Stack elements :", ob.display())
print("Peek element :", ob.peek())
print("Delete element :", ob.pop())
print("Delete element :", ob.pop())
print("Stack elements :", ob.display())
```

```python
# fixed size queue
class Queue :
    def __init__(self, n):
        self.max_size = n
        self.queue = [None] * self.max_size
        self.rear= -1
        self.front = -1
        
    def enqueue(self,val):
        if self.rear == self.max_size-1:
            return "Queue is full"
        if self.front == -1:
            self.front = 0
        self.rear +=1
        self.queue[self.rear] = val
        
    def dequeue(self):
        if self.front == -1:
            return "Queue is empty"
        temp = self.queue[self.front]
        self.queue[self.front] = None
        self.front +=1
        if self.rear == self.front:
            self.front = -1
            self.rear = -1
        return temp
        
    def getfront(self):
        if self.front == -1:
            if self.front == -1:
                return "Queue is empty"
        return self.queue[self.front]
    def getrear(self):
        if self.rear == -1:
            return "Queue is empty"
        return self.queue[self.rear]
    def display(self):
        if self.front == -1:
            return "Queue is empty"
        return self.queue
        
q=Queue(5)
q.enqueue(11)
q.enqueue(12)
q.enqueue(13)
q.enqueue(14)
print("queue : ",q.display()) 
print("front : ",q.getfront())
print("Rear : ",q.getrear())
print("Delete : ",q.dequeue()) 
print("queue : ",q.display()) 
```

```python
# implement the circular queue

class CircularQueue :
    def __init__(self, n):
        self.max_size = n
        self.queue = [None] * self.max_size
        self.rear= -1
        self.front = -1
        
    def enqueue(self,val):
        if (self.rear+1)%self.max_size == self.front:
            return "Queue is full"
        if self.front == -1:
            self.front = 0
        self.rear = (self.rear+1) % self.max_size
        self.queue[self.rear] = val
        
    def dequeue(self):
        if self.front == -1:
            return "Queue is empty"
        temp = self.queue[self.front]
        if self.front == self.rear:
            self.front = self.rear = -1
        else:
            self.front = (self.front+1) % self.max_size
        return temp
        
    def getfront(self):
        if self.front == -1:
            if self.front == -1:
                return "Queue is empty"
        return self.queue[self.front]
    def getrear(self):
        if self.rear == -1:
            return "Queue is empty"
        return self.queue[self.rear]
        
    def display(self):
        if self.front == -1:
            return "Queue is empty"
        result = []
        i= self.front
        while True:
            result.append(self.queue[i])
            if i == self.rear:
                break
            i = (i+1)% self.max_size       
        return result
        
q=CircularQueue(5)
q.enqueue(11)
q.enqueue(12)
q.enqueue(13)
q.enqueue(14)
print("queue : ",q.display()) 
print("front : ",q.getfront())
print("Rear : ",q.getrear())
print("Delete : ",q.dequeue()) 
print("queue : ",q.display()) 
```
